```python
import pandas as pd
```

```python
df = pd.read_csv("representation in movies.csv")
```

```python
df = df.drop(['positive representation','Unnamed: 2', 'index'], axis=1)
```

```python
ms = df[df['scale'] == 'mainstream']
```

```python
ms = ms.rename(columns={
    'diverse representation (dynamic character or other dynamic characters)':'diverse representation'})
```

```python
cha = []
for v in ms['characters']:
    each_row = [v]
    cha.append(each_row)
cha
```

```python
ms['character'] = cha
```

```python
ms = ms.explode('character')
```

```python
ms['character_2'] = [s.split(',') for s in ms['character']]
```

```python
ms = ms.explode('character_2')
```

```python
ms = ms.drop(['characters', 'character'], axis=1)
```

```python
ms = ms.rename(columns={
    'character_2' : 'characters',
    'villan' : 'villain'
})
```

```python
ms_villain = ms[ms['villain'] == 'yes']
```
